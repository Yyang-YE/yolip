Visualize

```python
import cv2
import matplotlib.pyplot as plt
img_name = image_path.split('/')[3]

#detect.py & detect2.py의 결과를 참조하여 경로 수정

ori = cv2.imread('/content/yolov5/runs/detect/exp/{0}'.format(img_name))  # use coco image
#ori = cv2.imread('/content/yolov5/runs/detect/exp5/011.jpg'.format(img_name), cv2.IMREAD_COLOR)  # use personal image
ori = cv2.cvtColor(ori,cv2.COLOR_BGR2RGB)

out = cv2.imread('/content/yolov5/runs/detect/exp2/{0}'.format(img_name))  # use coco image
#out = cv2.imread('/content/yolov5/runs/detect/exp9/011.jpg'.format(img_name))  # use personal image
out = cv2.cvtColor(out,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(ori)
plt.title('Original Image(YOLO)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(out)
plt.title('Processed Image(YOLO + CLIP)')
plt.axis('off')

plt.show()
```
